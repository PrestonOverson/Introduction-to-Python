```python
master_df = pd.read_csv('kc_house_data.csv')
```

```python
master_df.head()
```

```python
data = master_df.drop(columns = ['lat', 'long'])
yr_renov_bool = master_df['yr_renovated'].copy()
j = 0
for i in yr_renov_bool:
    if yr_renov_bool[j] > 1:
        yr_renov_bool[j] = 1
    j += 1
data['renovated'] = yr_renov_bool
data = data.drop(columns = ['yr_renovated'])
```

```python
data.info()
```

```python
dates = data['date'].copy()
months = []
for i in dates:
    mon = i[0:6]
    months.append(mon)
data['date'] = months
```

```python
data = pd.get_dummies(data, columns=['date'])
```

```python
data.info()
```

```python
data.to_csv(r'kc_housing_clean.csv', index = 0)
```
